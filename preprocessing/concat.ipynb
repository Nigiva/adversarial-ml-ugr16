```python
EXTRACT_PATH = "../data/extract/"
SAMPLES_PATH = os.path.join(EXTRACT_PATH, "samples")
HEADER = ["datetime_end", "duration", "source_ip", "destination_ip", "source_port", "destination_port", "protocol", "flag", "forwarding", "type",
          "packet", "byte", "label"]
LABEL_LIST = ["background", "blacklist", "anomaly-spam", "anomaly-sshscan", "anomaly-udpscan", "dos", "scan11", "scan44", "nerisbotnet"]
```

## Concat samples

```python
samples_path_list = glob.glob(os.path.join(SAMPLES_PATH, "*.csv"))
samples_path_list
```

```python
dataset_df = pd.read_csv(samples_path_list[0], names=HEADER)
for samples_path in tqdm(samples_path_list[1:]):
    sample_df = pd.read_csv(samples_path, names=HEADER)
    dataset_df = pd.concat([dataset_df, sample_df])
```

```python
dataset_df
```

```python
dataset_df["label"].value_counts()
```

```python
900_000*8
```

```python
DATASET_LARGE_NAME = "ugr16-large.csv"
export_path = os.path.join(EXTRACT_PATH, DATASET_LARGE_NAME)
dataset_df.to_csv(export_path)
```

## Reduce dataset

### Base

```python
dataset_df = pd.read_csv(os.path.join(EXTRACT_PATH, "ugr16-large.csv"), index_col=0)
```

```python
dataset_df
```

```python
dataset_df["label"].value_counts()
```

```python
def sample_label(df, sample_size=90_000):
    if (df.label == "background").all():
        return df.sample(n=sample_size*8)
    return df.sample(n=sample_size)
```

```python
reduced_dataset_df = dataset_df.groupby("label").progress_apply(sample_label).reset_index(drop=True)
```

```python
reduced_dataset_df
```

```python
reduced_dataset_df["label"].value_counts()
```

```python
DATASET_BASE_NAME = "ugr16-base.csv"
export_path = os.path.join(EXTRACT_PATH, DATASET_BASE_NAME)
reduced_dataset_df.to_csv(export_path)
```
